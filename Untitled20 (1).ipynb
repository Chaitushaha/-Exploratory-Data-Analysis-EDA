```python
#step 1 : load libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# step 2: load dataset
df = pd.read_csv("C:\\Users\\NMC\\Downloads\\netflix_titles.csv")
```

```python
# step 3 : data cleaning first five rows 
df.head()
```

```python
# Missing values count
missing_before = df.isnull().sum()
print(missing_before)
```

```python
# Missing categorical values were handled by replacing them with “Unknown”.
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])
```

```python
# Unnecessary columns are removed for better analysis.
df = df.dropna(subset=['date_added'])
df = df.dropna(subset=['duration'])
```

```python
#rechecking missing values
df.isnull().sum()
```

**Data Cleaning Summary**:
The Netflix dataset contained missing values in columns like director, cast, country, rating, and date_added.
Categorical columns such as director, cast, and country were filled with “Unknown”.
Missing values in the rating column were handled using mode imputation.
Rows with missing date_added and duration were removed.
After cleaning, the dataset was verified and prepared for Exploratory Data Analysis.

**Exploratory Data Analysis**

```python
# step 1 : import library
import matplotlib.pyplot as plt
```

```python
# step 2 : 
#Plot distribution of numerical features using histograms
plt.figure(figsize=(8,5))
plt.hist(df['release_year'], bins=20)
plt.xlabel('Release Year')
plt.ylabel('Number of Movies/TV Shows')
plt.title('Distribution of Release Year on Netflix')
plt.show()
```

This graph shows how many titles were released each year. Most titles were released after 2015.
This means Netflix added more content in recent years.

```python
# step 3 : 
# Analyze categorical features using count plots.
df['type'].value_counts().plot(kind='bar')
plt.xlabel('Type')
plt.ylabel('Count')
plt.title('Count of Movies vs TV Shows on Netflix')
plt.show()
```

This graph shows the number of Movies and TV Shows. Movies are more than TV Shows.
This means Netflix has more Movies than TV Shows.

```python
df['rating'].value_counts().plot(kind='bar')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.title('Content Rating Distribution on Netflix')
plt.show()
```

This graph shows the distribution of content ratings.
TV-MA and TV-14 ratings are the most common.
This means most content is made for mature audiences.

```python
# step 4 : Use box plots to identify outliers
plt.figure(figsize=(6,4))
plt.boxplot(df['release_year'])
plt.ylabel('Release Year')
plt.title('Box Plot of Release Year')
plt.show()
```

This graph shows the spread of content duration.
Most titles have similar duration.
A few titles have very high duration, which are outliers.

```python
# step 5 : Plot correlation heatmap to understand feature relationships
numeric_df = df.select_dtypes(include='number')
corr = numeric_df.corr()

plt.figure(figsize=(4,3))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

This graph shows the relationship between numerical features.
There is no strong correlation between most features.
The features behave independently.
